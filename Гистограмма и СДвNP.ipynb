# ГИСТОГРАММА

```python
np.random.seed(42)
x=np.random.randn(1000)
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x)
```

```python
bins=np.linspace(-5,5,20)
bins
```

```python
counts=np.zeros_like(bins)
counts
```

```python
i=np.searchsorted(bins,x)
i
```

```python
np.add.at(counts,i,1)
counts
```

```python
plt.plot(bins,counts)
```

```python
plt.step(bins,counts)
```

# Структурированные массивы в NumPy

```python
name = ['Ildar', 'Akvina', 'Daniar', 'Diana']
age = [34, 34, 8, 2]
weight = [71, 51, 30, 13]
```

```python
data=np.zeros(4, dtype={'names':('name', 'age', 'weight'), 'formats':('U10', 'i4', 'f8')})
```

```python
data.dtype
```

```python
data
```

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
data
```

```python
data[0]
```

```python
data[data['age']<30]
```

```python
data[data['age']<30]['weight'].mean()
```
